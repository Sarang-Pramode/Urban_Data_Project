#  <b>Population and Demographics</b>
## (a) Decennial Census

```python
# Decennial Census - Reading CSV files

Decenial_Census_2010_table_path = 'Decenial Census/DECENNIALPL2010.P1_data_with_overlays_2021-11-17T003605.csv'
Decenial_Census_2020_table_path = 'Decenial Census/DECENNIALPL2020.P1_data_with_overlays_2021-11-17T003605.csv'

Decenial_Census_2010_df = pd.read_csv(Decenial_Census_2010_table_path)
Decenial_Census_2020_df = pd.read_csv(Decenial_Census_2020_table_path)
```

```python
#Taking only Desired Columns

Decenial_Census_2010_Interested_Columns = ['NAME','P001001','P001003','P001004','P001006']
Decenial_Census_2020_Interested_Columns = ['NAME','P1_001N','P1_003N','P1_004N','P1_006N']
"""
yr:2010,2020

- NAME : Census Tract Name
- P001001,P1_001N : Total Population
- P001003,P1_003N : Total Population White Alone
- P001004,P1_004N : Total Population Black/African American Alone
- P001006,P1_006N : Total Population Asian Alone"""

Decenial_Census_2010_df_reduced = Decenial_Census_2010_df[Decenial_Census_2010_Interested_Columns]
Decenial_Census_2020_df_reduced = Decenial_Census_2020_df[Decenial_Census_2020_Interested_Columns]
```

## Preprocessing - Decennial Census tables

```python
#Renaming Columns - yr 2010
Decenial_Census_2010_df_reduced.rename(columns={'P001001':'Total_Population','P001003':'Total_White_Population',\
    'P001004':'Total_Black_Population','P001006':'Total_Asian_Population'},inplace=True)
#Renaming Columns - yr 2020
Decenial_Census_2020_df_reduced.rename(columns={'P1_001N':'Total_Population','P1_003N':'Total_White_Population',\
    'P1_004N':'Total_Black_Population','P1_006N':'Total_Asian_Population'},inplace=True)

#Splitting NAME to Census_tract, County and State
Decenial_Census_2020_df_reduced[['Census_Tract','County','State']] = Decenial_Census_2020_df_reduced['NAME'].str.split(',',expand=True)
Decenial_Census_2010_df_reduced[['Census_Tract','County','State']] = Decenial_Census_2010_df_reduced['NAME'].str.split(',',expand=True)

#Retaining only numbers in the Census tract column
Decenial_Census_2020_df_reduced.Census_Tract = Decenial_Census_2020_df_reduced.Census_Tract.str.replace(r'[a-zA-Z]','')
Decenial_Census_2010_df_reduced.Census_Tract = Decenial_Census_2010_df_reduced.Census_Tract.str.replace(r'[a-zA-Z]','')

#Remove leading and trailing spaces in census tract column
Decenial_Census_2010_df_reduced.Census_Tract = list(map(lambda x : x.replace(" ",""),Decenial_Census_2010_df_reduced.Census_Tract))
Decenial_Census_2020_df_reduced.Census_Tract = list(map(lambda x : x.replace(" ",""),Decenial_Census_2020_df_reduced.Census_Tract))
```

```python
Decenial_Census_2020_df_reduced.head()
```

---

```python
Decenial_Census_2010_final_df = Decenial_Census_2010_df_reduced
Decenial_Census_2020_final_df = Decenial_Census_2020_df_reduced
```

```python
#Filtering Columns
ACS_Demographic_and_Housing_interested_columns_2010 = ['NAME','DP05_0001E','DP05_0002E','DP05_0003E','DP05_0009PE','DP05_0010PE',\
    'DP05_0011PE','DP05_0017E']
ACS_Demographic_and_Housing_interested_columns_2019 = ['NAME','DP05_0001E','DP05_0002E','DP05_0003E','DP05_0010PE',\
    'DP05_0011PE','DP05_0012PE','DP05_0018E']

ACS_Demographic_and_Housing_2010_df_reduced = ACS_Demographic_and_Housing_2010_df[ACS_Demographic_and_Housing_interested_columns_2010]
ACS_Demographic_and_Housing_2019_df_reduced = ACS_Demographic_and_Housing_2019_df[ACS_Demographic_and_Housing_interested_columns_2019]

#Renaming Columns
ACS_Demographic_and_Housing_2010_df_reduced.rename(columns={'DP05_0001E':'Total_Population','DP05_0002E':'Total_Male_Population',\
    'DP05_0003E':'Total_Female_Population',\
        'DP05_0009PE':'Per_Population_25_to_34_yrs',\
        'DP05_0010PE':'Per_Population_35_to_44_yrs',\
        'DP05_0011PE':'Per_Population_45_to_54_yrs',\
            'DP05_0017E':'Median_Age'},inplace=True)

#Renaming Columns
ACS_Demographic_and_Housing_2019_df_reduced.rename(columns={'DP05_0001E':'Total_Population','DP05_0002E':'Total_Male_Population',\
    'DP05_0003E':'Total_Female_Population',\
        'DP05_0010PE':'Per_Population_25_to_34_yrs',\
        'DP05_0011PE':'Per_Population_35_to_44_yrs',\
        'DP05_0012PE':'Per_Population_45_to_54_yrs',\
            'DP05_0018E':'Median_Age'},inplace=True)

#Splitting NAME to Census_tract, County and State
ACS_Demographic_and_Housing_2010_df_reduced[['Census_Tract','County','State']] = ACS_Demographic_and_Housing_2010_df_reduced['NAME'].str.split(',',expand=True)
ACS_Demographic_and_Housing_2019_df_reduced[['Census_Tract','County','State']] = ACS_Demographic_and_Housing_2019_df_reduced['NAME'].str.split(',',expand=True)

#Retaining only numbers in the Census tract column
ACS_Demographic_and_Housing_2010_df_reduced.Census_Tract = ACS_Demographic_and_Housing_2010_df_reduced.Census_Tract.str.replace(r'[a-zA-Z]','')
ACS_Demographic_and_Housing_2019_df_reduced.Census_Tract = ACS_Demographic_and_Housing_2019_df_reduced.Census_Tract.str.replace(r'[a-zA-Z]','')

#Remove leading and trailing spaces in census tract column
ACS_Demographic_and_Housing_2019_df_reduced.Census_Tract = list(map(lambda x : x.replace(" ",""),ACS_Demographic_and_Housing_2019_df_reduced.Census_Tract))
ACS_Demographic_and_Housing_2010_df_reduced.Census_Tract = list(map(lambda x : x.replace(" ",""),ACS_Demographic_and_Housing_2010_df_reduced.Census_Tract))
```
